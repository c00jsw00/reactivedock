```
import csv
import py3Dmol
import os
from openbabel import pybel
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import MolToImage
#
with open('database.csv', 'r', newline='') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        parameter1, parameter2, parameter3, parameter4, parameter5, parameter6  = row[0], row[1], row[2], row[3], row[4], row[5]
        namesdf  = parameter1+'.sdf'
        namepdbqt = parameter1+'.pdbqt'
        smiles = parameter2
        mol = Chem.MolFromSmiles(smiles)
        smi = Chem.MolToSmiles(mol)
        out=pybel.Outputfile(filename=namepdbqt,format='pdbqt',overwrite=True)
        pdbqt = pybel.readstring("smi", smi)
        pdbqt.make3D(forcefield='mmff94', steps=100)
        out.write(pdbqt)
        out.close()
```
